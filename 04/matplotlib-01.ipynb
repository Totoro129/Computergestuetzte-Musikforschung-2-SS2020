## matplotlib (pyplot) einbinden

```python
from matplotlib import pyplot as plt
```

## hilfsfunktion...

```python
vals = [
    (1, 3),
    (20, 4),
    (3, 9),
    (5, 7)
]

def transform(vals):
    x = []
    y = []
    for val in vals:
        x.append(val[0])
        y.append(val[1])
    return x, y

x, y = transform(vals)
print(x, y)
```

## Einfache Plots
### Linie zeichnen
* benötigt x- und y-Koordinaten als separate Listen

```python
# line plotting
x = [5, 2, 9, 4, 7]
y = [10, 5, 8 , 4 ,2]
plt.plot(x, y)
```

## Bardiagramm
* benötigt x- und y-Koordinaten als separate Listen

```python
# bar plotting
plt.bar(x, y)
```

## Histogramm

```python
# histogram plotting
y = [13, 1, 2, 6, 4]
plt.hist(y)
```

## Streudiagramm
* benötigt x- und y-Koordinaten als separate Listen

```python
# scatter plotting
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x,y,'ro')
# ro: red circle
```

## Achsen anpassen
* `plt.axis()` nimmt eine Liste mit Koordinaten: `[x-min, x-max, y-min, y-max]`

```python
# scatter plotting
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x,y,'ro')
# ro: rote Kreise
# gx: grüne Kreuze
# y--: gelb gestrichelt

plt.axis([0, 6, 0, 20])
```

```python
# line plotting
x = [5, 2, 9, 4, 7]
y = [10, 5, 8 , 4 ,2]
plt.plot(x, y)

plt.axis([-10, 10, 0, 30])
```
