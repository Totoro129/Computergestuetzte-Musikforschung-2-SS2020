```python
sinus = np.array([])

nsteps = 44100
for i in range(nsteps + 1):
    sinus = np.append(sinus, np.sin( i / nsteps * 2 * np.pi * 440))
#     print(i / nsteps * 2 * np.pi) # 0 -> 2 * pi

plt.plot(sinus)
plt.show()
```

# IPython.display

```python
import IPython.display as ipd

ipd.Audio(sinus, rate=44100)
```

```python
length = 3 # in Sekunden
sample_rate = 44100
freq = 50
sinus = np.sin(np.linspace(0, 2 * np.pi * freq * length, sample_rate * length))
plt.figure(figsize=(18, 6))
plt.plot(sinus)
ipd.Audio(sinus, rate=sample_rate)
```
